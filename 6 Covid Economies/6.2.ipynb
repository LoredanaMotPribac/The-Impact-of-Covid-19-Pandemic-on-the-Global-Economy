```python
#Q6
#Creating a categorical variable that splits the "GDPCAP" column into categories
sns.distplot(df['GDPCAP'], bins = 20)
```

```python
#Creating a categorical variable that splits the "TD" column into categories
sns.distplot(df['TD'], bins = 20)
```

```python
df.loc[df['GDPCAP'] < 7, 'Country GDP'] = 'Poor country'
```

```python
df.loc[df['GDPCAP'] >= 9, 'Country GDP'] = 'Rich country'
```

```python
df.loc[(df['GDPCAP'] >= 7) & (df['GDPCAP'] < 9), 'Country GDP'] = 'Middle country'
```

```python
df['Country GDP'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the GDPCAP categories created above
sns.set(style="ticks")
g = sns.catplot(x="Population", y="GDPCAP", hue="Country GDP", data=df)
```

```python
df.loc[df['TD'] >= 9, 'Country Covid'] = 'Low deaths'
```

```python
df.loc[df['TD'] <= 1, 'Country Covid'] = 'High deaths'
```

```python
df.loc[(df['TD'] > 1) & (df['TD'] < 9), 'Country Covid'] = 'Middle deaths'
```

```python
df[ 'Country Covid'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the TD categories created above
sns.set(style="ticks")
g = sns.catplot(x="Population", y="TD", hue="Country Covid", data=df)
```
